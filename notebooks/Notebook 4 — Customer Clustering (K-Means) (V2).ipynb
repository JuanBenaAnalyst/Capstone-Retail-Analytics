```python
# ============================================
# Notebook — Customer Clustering (GPU/CPU)
# Blocks 0–3 (English only, with fixes)
# ============================================

# ---------- Block 0: Setup & Backend ----------
# Set threading limits BEFORE importing BLAS-backed libs
import os
os.environ.setdefault("OMP_NUM_THREADS", "16")
os.environ.setdefault("OPENBLAS_NUM_THREADS", "16")
os.environ.setdefault("MKL_NUM_THREADS", "16")
os.environ.setdefault("VECLIB_MAXIMUM_THREADS", "16")
os.environ.setdefault("NUMEXPR_NUM_THREADS", "16")

import warnings, json, time, hashlib
from pathlib import Path
from datetime import datetime
from typing import Optional
import numpy as np
import pandas as pd

from sklearn.metrics import (
    silhouette_score as sk_silhouette,
    davies_bouldin_score,
    calinski_harabasz_score,
)

RANDOM_SEED = 42
np.random.seed(RANDOM_SEED)
pd.set_option("display.float_format", lambda x: f"{x:,.4f}")
pd.set_option("display.max_columns", 120)
warnings.filterwarnings("ignore")

# Backend selector (RAPIDS if available)
BACKEND = os.environ.get("CLUSTER_BACKEND", "gpu").lower()
_gpu_ok = False
_cuml = _cudf = None

if BACKEND == "gpu":
    try:
        import cuml, cudf
        from cuml.cluster import KMeans as cuKMeans  # noqa: F401
        _cuml, _cudf, _gpu_ok = cuml, cudf, True
    except Exception as e:
        print("[WARN] RAPIDS/cuML not available. Falling back to CPU. Detail:", repr(e))
        BACKEND = "cpu"

# --- Unified adapters (KMeans + Silhouette + PCA)
from sklearn.cluster import KMeans as skKMeans

def _to_numpy32(X):
    """Return a float32 numpy array (no copy if not needed)."""
    if isinstance(X, pd.DataFrame):
        return X.to_numpy(dtype=np.float32, copy=False)
    X = np.asarray(X)
    return X.astype(np.float32, copy=False)

class KMeansAdapter:
    def __init__(self, n_clusters:int, n_init:int=20, max_iter:int=300, random_state:int=RANDOM_SEED):
        self.n_clusters, self.n_init, self.max_iter, self.random_state = n_clusters, n_init, max_iter, random_state
        self.model = None
        self.inertia_ = None
        self.backend = "gpu" if (BACKEND == "gpu" and _gpu_ok) else "cpu"

    def fit_predict(self, X):
        if self.backend == "gpu":
            from cuml.cluster import KMeans as cuKMeans  # local import (keeps import lazy)
            X32 = _to_numpy32(X)
            self.model = cuKMeans(n_clusters=self.n_clusters, n_init=self.n_init,
                                  max_iter=self.max_iter, random_state=self.random_state, output_type="numpy")
            labels = self.model.fit_predict(X32)
            # Manually compute inertia/SSE when not exposed by cuML
            try:
                centers = self.model.cluster_centers_
                dif = X32 - centers[labels]
                self.inertia_ = float(np.sum(dif * dif))
            except Exception:
                self.inertia_ = None
            return labels
        else:
            self.model = skKMeans(n_clusters=self.n_clusters, n_init=self.n_init,
                                  max_iter=self.max_iter, random_state=self.random_state)
            labels = self.model.fit_predict(X)
            self.inertia_ = float(self.model.inertia_)
            return labels

# Silhouette GPU→CPU fallback
try:
    from cuml.metrics import silhouette_score as cu_silhouette
except Exception:
    cu_silhouette = None

def silhouette_score_adapter(X, labels):
    if BACKEND == "gpu" and cu_silhouette is not None:
        try:
            return float(cu_silhouette(_to_numpy32(X), labels))
        except Exception:
            pass
    return float(sk_silhouette(np.asarray(X, dtype=np.float32), labels))

# PCA adapter (cuML PCA has no random_state)
try:
    from cuml.decomposition import PCA as cuPCA
except Exception:
    cuPCA = None
from sklearn.decomposition import PCA as skPCA

class PCAAdapter:
    def __init__(self, n_components=2, random_state:int=RANDOM_SEED):
        if BACKEND == "gpu" and cuPCA is not None:
            self.model = cuPCA(n_components=n_components)  # no random_state arg
        else:
            self.model = skPCA(n_components=n_components, random_state=random_state)
    def fit_transform(self, X):
        return self.model.fit_transform(X)
    @property
    def explained_variance_ratio_(self):
        return getattr(self.model, "explained_variance_ratio_", None)

# Utilities
def _md5_cols(cols:list[str]) -> str:
    return hashlib.md5((",").join(cols).encode("utf-8")).hexdigest()[:8]

def log_environment():
    print("=== Environment ===")
    print("Backend:", BACKEND, "| GPU available:", _gpu_ok)
    try:
        import platform
        print("Python:", platform.python_version())
    except Exception:
        pass
    if BACKEND == "gpu":
        try:
            import cuml, cudf
            print("cuML:", cuml.__version__, "| cuDF:", cudf.__version__)
        except Exception as e:
            print("RAPIDS versions not available:", repr(e))
    else:
        import sklearn, numpy
        print("NumPy:", numpy.__version__, "| scikit-learn:", sklearn.__version__)
    print("===================")

log_environment()

# ---------- Block 1: Load & Sanity ----------
from IPython.display import display

def _log(msg: str):
    print(f"[{datetime.now().strftime('%Y-%m-%d %H:%M:%S')}] {msg}")

DATA_DIR_IN = Path("clean/model_input")
PATH_FULL   = DATA_DIR_IN / "model_features_customers_full.parquet"
PATH_RFM3   = DATA_DIR_IN / "model_features_customers_rfm3.parquet"
PATH_FCAT   = DATA_DIR_IN / "feature_catalog.json"

assert PATH_FULL.exists(), f"Missing {PATH_FULL}"
assert PATH_RFM3.exists(), f"Missing {PATH_RFM3}"

_log("Loading feature matrices…")
df_full = pd.read_parquet(PATH_FULL)
df_rfm3 = pd.read_parquet(PATH_RFM3)
_log(f"FULL: rows={len(df_full):,}, cols={df_full.shape[1]}")
_log(f"RFM3: rows={len(df_rfm3):,}, cols={df_rfm3.shape[1]}")

id_col = "CustomerID"
assert id_col in df_full.columns and id_col in df_rfm3.columns, "CustomerID is required in both matrices."

# Use feature catalog if present; fallback to *_scaled columns
if PATH_FCAT.exists():
    with open(PATH_FCAT, "r", encoding="utf-8") as f:
        fcat = json.load(f)
    feat_full = [c for c in fcat.get("features_full", []) if c in df_full.columns]
    feat_rfm3 = [c for c in fcat.get("features_rfm3", []) if c in df_rfm3.columns]
else:
    feat_full = [c for c in df_full.columns if c.endswith("_scaled")]
    feat_rfm3 = [c for c in df_rfm3.columns if c.endswith("_scaled")]

assert len(feat_rfm3) == 3, f"Expected exactly 3 RFM3 scaled features; got {len(feat_rfm3)}."
assert len(feat_full) >= 5, f"Expected ≥5 scaled features in FULL; got {len(feat_full)}."

# Basic checks
checks = []
def add_check(name, cond): checks.append({"Check": name, "Pass": bool(cond)})

add_check("≥ 1,000 customers (FULL)", len(df_full) >= 1_000)
add_check("CustomerID not null (FULL)", df_full[id_col].isna().sum() == 0)
add_check("1 row per CustomerID (FULL)", df_full[id_col].nunique() == len(df_full))
add_check("CustomerID not null (RFM3)", df_rfm3[id_col].isna().sum() == 0)
add_check("1 row per CustomerID (RFM3)", df_rfm3[id_col].nunique() == len(df_rfm3))

nan_rates_full = df_full[feat_full].isna().mean()
nan_rates_rfm3 = df_rfm3[feat_rfm3].isna().mean()
add_check("No NaNs in FULL features", float(nan_rates_full.max()) == 0.0)
add_check("No NaNs in RFM3 features", float(nan_rates_rfm3.max()) == 0.0)

eps = 1e-9
mins_full = df_full[feat_full].min(); maxs_full = df_full[feat_full].max()
mins_rfm3 = df_rfm3[feat_rfm3].min(); maxs_rfm3 = df_rfm3[feat_rfm3].max()
add_check("FULL features within [0,1]", (mins_full >= -eps).all() and (maxs_full <= 1+eps).all())
add_check("RFM3 features within [0,1]", (mins_rfm3 >= -eps).all() and (maxs_rfm3 <= 1+eps).all())

print("\n=== Block 1 — Sanity Report ===")
print(f"- ID column: {id_col}")
print(f"- FULL feature set ({len(feat_full)}): {feat_full}")
print(f"- RFM3 feature set ({len(feat_rfm3)}): {feat_rfm3}")

print("\nGo/No-Go checks:")
display(pd.DataFrame(checks))

print("\nNaN rates (FULL):")
display(nan_rates_full.round(6).to_frame("nan_rate").T)

print("Bounds (FULL): min/max per feature")
display(pd.DataFrame({"min": mins_full.round(6), "max": maxs_full.round(6)}).T)

print("\nNaN rates (RFM3):")
display(nan_rates_rfm3.round(6).to_frame("nan_rate").T)

print("Bounds (RFM3): min/max per feature")
display(pd.DataFrame({"min": mins_rfm3.round(6), "max": maxs_rfm3.round(6)}).T)

# Hand-off: matrices + IDs (float32 if GPU)
X_full = df_full[feat_full].copy()
X_rfm3 = df_rfm3[feat_rfm3].copy()
if BACKEND == "gpu":
    X_full = X_full.astype(np.float32)
    X_rfm3 = X_rfm3.astype(np.float32)

cust_ids_full = df_full[[id_col]].copy()
cust_ids_rfm3 = df_rfm3[[id_col]].copy()

print("\n✅ Block 1 OK — proceed to the quick K-scan.")

# ---------- Block 2A: Fast K Scan ----------
import matplotlib.pyplot as plt

USE_SAMPLE      = True
SAMPLE_N        = 20_000
K_RANGE         = range(2, 11)   # 2..10
N_INIT_QUICK    = 10
MAX_ITER        = 300

def _log(msg):
    print(f"[{datetime.now().strftime('%Y-%m-%d %H:%M:%S')}] {msg}")

# Build CustomerID-indexed frames for safe sampling/selection
X_full_id = X_full.join(cust_ids_full).set_index(id_col)
X_rfm3_id = X_rfm3.join(cust_ids_rfm3).set_index(id_col)
common_ids = X_full_id.index.intersection(X_rfm3_id.index)

if USE_SAMPLE and len(common_ids) > SAMPLE_N:
    sampled_ids = (
        pd.Series(common_ids, name=id_col)
        .sample(n=SAMPLE_N, random_state=RANDOM_SEED)
        .tolist()
    )
    Xf_scan = X_full_id.loc[sampled_ids]
    Xr_scan = X_rfm3_id.loc[sampled_ids]
    _log(f"Sampling enabled → using {len(Xf_scan):,} rows for the scan.")
else:
    Xf_scan = X_full_id.loc[common_ids]
    Xr_scan = X_rfm3_id.loc[common_ids]
    _log(f"Sampling disabled → using full {len(Xf_scan):,} rows.")

def evaluate_kmeans_once(X, k, n_init=N_INIT_QUICK, max_iter=MAX_ITER):
    km = KMeansAdapter(n_clusters=k, n_init=n_init, max_iter=max_iter, random_state=RANDOM_SEED)
    t0 = time.perf_counter()
    labels = km.fit_predict(X)
    fit_s = time.perf_counter() - t0

    t1 = time.perf_counter()
    sil = silhouette_score_adapter(X, labels)
    sil_s = time.perf_counter() - t1

    X_cpu = np.asarray(X, dtype=np.float32)
    db = float(davies_bouldin_score(X_cpu, labels))
    ch = float(calinski_harabasz_score(X_cpu, labels))

    return {
        "k": int(k),
        "inertia": km.inertia_,
        "silhouette": float(sil),
        "davies_bouldin": db,
        "calinski_harabasz": ch,
        "fit_s": fit_s,
        "silhouette_s": sil_s,
        "backend": BACKEND,
    }

def grid_scan(X, label):
    _log(f"Scanning {label} …")
    rows = []
    for k in K_RANGE:
        rows.append(evaluate_kmeans_once(X, k))
    dfk = pd.DataFrame(rows).sort_values("k").reset_index(drop=True)

    print(f"\n=== {label} — Quick scan results ===")
    display(dfk[["k","silhouette","davies_bouldin","calinski_harabasz","fit_s","silhouette_s","inertia","backend"]])

    fig, axes = plt.subplots(1, 2, figsize=(14, 5))
    if dfk["inertia"].notna().any():
        axes[0].plot(dfk["k"], dfk["inertia"], marker="o")
        axes[0].set_title(f"Elbow (Inertia) — {label}")
        axes[0].set_xlabel("k"); axes[0].set_ylabel("Inertia")
    else:
        axes[0].text(0.5, 0.5, "Inertia unavailable on GPU path", ha="center", va="center")
        axes[0].set_axis_off()
    axes[1].plot(dfk["k"], dfk["silhouette"], marker="o")
    axes[1].set_title(f"Silhouette — {label}")
    axes[1].set_xlabel("k"); axes[1].set_ylabel("Silhouette")
    plt.tight_layout(); plt.show()

    best_k = int(dfk.iloc[dfk["silhouette"].idxmax()]["k"])
    print(f"[Quick suggestion] {label} best k by silhouette: {best_k}")
    return dfk, best_k

df_full_quick, best_full_quick = grid_scan(Xf_scan, "FULL (quick)")
df_rfm3_quick, best_rfm3_quick = grid_scan(Xr_scan, "RFM3 (quick)")

print("\n✅ Block 2A complete")

# ---------- Block 2B: Stable confirmation on FULL data ----------
from IPython.display import display

RANDOM_SEED   = 42
N_INIT_STABLE = 50
MAX_ITER      = 300
K_MIN, K_MAX  = 2, 10

def candidate_list_from_scan(df_quick: pd.DataFrame, kmin=K_MIN, kmax=K_MAX):
    """Use best silhouette ± 1 (within bounds) as candidates."""
    best_k = int(df_quick.iloc[df_quick["silhouette"].idxmax()]["k"])
    cands = {best_k}
    if best_k - 1 >= kmin: cands.add(best_k - 1)
    if best_k + 1 <= kmax: cands.add(best_k + 1)
    return sorted(cands), best_k

cand_full, best_full_quick = candidate_list_from_scan(df_full_quick)
cand_rfm3, best_rfm3_quick = candidate_list_from_scan(df_rfm3_quick)

print(f"Candidates (FULL) from quick scan (best={best_full_quick}): {cand_full}")
print(f"Candidates (RFM3) from quick scan (best={best_rfm3_quick}): {cand_rfm3}")

def confirm_grid(X: pd.DataFrame, cands: list[int], label: str) -> tuple[pd.DataFrame, int]:
    rows = []
    _log(f"Confirming {label} on FULL data with n_init={N_INIT_STABLE} …")
    for k in cands:
        km = KMeansAdapter(n_clusters=int(k), n_init=N_INIT_STABLE, max_iter=MAX_ITER, random_state=RANDOM_SEED)
        labels = km.fit_predict(X)
        sil = float(silhouette_score_adapter(X, labels))
        X_cpu = np.asarray(X, dtype=np.float32)
        db = float(davies_bouldin_score(X_cpu, labels))
        ch = float(calinski_harabasz_score(X_cpu, labels))
        rows.append({"k": int(k), "inertia": km.inertia_, "silhouette": sil,
                     "davies_bouldin": db, "calinski_harabasz": ch})
    out = pd.DataFrame(rows).sort_values("silhouette", ascending=False).reset_index(drop=True)
    display(out)
    final_k = int(out.iloc[0]["k"])
    print(f"[Final pick] {label} → k = {final_k}")
    return out, final_k

df_full_confirm, final_k_full = confirm_grid(X_full, cand_full, "FULL")
df_rfm3_confirm, final_k_rfm3 = confirm_grid(X_rfm3, cand_rfm3, "RFM3")

# --- Export quick/confirm scans (make Block 2 self-contained)
from pathlib import Path

try:
    OUT_DIR  # already defined in Block 3?
except NameError:
    DATA_DIR_IN = Path("clean/model_input")
    OUT_DIR = DATA_DIR_IN / "clusters"

SCAN_DIR = OUT_DIR / "scan"
SCAN_DIR.mkdir(parents=True, exist_ok=True)

# Save quick scan tables
df_full_quick.to_csv(SCAN_DIR / "k_scan_full_quick.csv", index=False)
df_rfm3_quick.to_csv(SCAN_DIR / "k_scan_rfm3_quick.csv", index=False)

# Save confirmation tables (final candidates)
df_full_confirm.to_csv(SCAN_DIR / "k_confirm_full.csv", index=False)
df_rfm3_confirm.to_csv(SCAN_DIR / "k_confirm_rfm3.csv", index=False)

print(f"[OK] Scan exports → {SCAN_DIR}")



print("\n✅ Block 2B complete — if you agree with these k values, run Block 3.")

# ---------- Block 3: Train, Label, Profile, PCA, Export (with FIX) ----------
import matplotlib.pyplot as plt

DATA_DIR_IN = Path("clean/model_input")
OUT_DIR     = DATA_DIR_IN / "clusters"
OUT_DIR.mkdir(parents=True, exist_ok=True)

def _now(): return datetime.now().strftime("%H:%M:%S")

def fit_kmeans_and_report(X: pd.DataFrame, k: int, label: str, n_init:int=50, max_iter:int=300):
    print(f"[{_now()}] Training KMeans ({label}) with k={k}, n_init={n_init} …")
    km = KMeansAdapter(n_clusters=k, n_init=n_init, max_iter=max_iter, random_state=RANDOM_SEED)
    labels = km.fit_predict(X)
    sil = float(silhouette_score_adapter(X, labels))
    X_cpu = np.asarray(X, dtype=np.float32)
    db  = float(davies_bouldin_score(X_cpu, labels))
    ch  = float(calinski_harabasz_score(X_cpu, labels))
    print(f"  → {label}: silhouette={sil:.4f}, davies_bouldin={db:.4f}, calinski_harabasz={ch:,.1f}")
    return km, labels, sil, db, ch

def attach_labels(cust_ids_df: pd.DataFrame, labels: np.ndarray, colname: str) -> pd.DataFrame:
    out = cust_ids_df.copy(); out[colname] = labels; return out

# ===== FIX: robust cluster_profile (flatten without losing the segment column name)
def cluster_profile(df: pd.DataFrame, cluster_col: str) -> pd.DataFrame:
    feats = [c for c in df.columns if c.endswith("_scaled")]
    agg = df.groupby(cluster_col, as_index=False)[feats]\
            .agg(["mean","median","std","min","max","count"]).reset_index()

    new_cols = []
    for col in agg.columns.to_flat_index():
        if isinstance(col, tuple):
            base, stat = col
            new_cols.append(f"{base}_{stat}" if stat else base)
        else:
            new_cols.append(col)
    agg.columns = new_cols

    sizes = df[cluster_col].value_counts().sort_index().rename("count")
    agg = agg.merge(sizes, left_on=cluster_col, right_index=True)
    agg["share"] = agg["count"] / len(df)
    return agg
# ===== END FIX

def executive_profile(profile_df: pd.DataFrame, cluster_col: str) -> pd.DataFrame:
    keep_meds = [
        "recency_days_scaled_median",
        "frequency_scaled_median",
        "monetary_scaled_median",
        "avg_ticket_scaled_median",
        "pct_discounted_scaled_median",
        "n_categories_scaled_median",
    ]
    keep = [c for c in keep_meds if c in profile_df.columns]
    slim = profile_df[[cluster_col, *keep, "share", "count"]].copy()
    num_cols = [c for c in slim.columns if c != cluster_col]
    slim[num_cols] = slim[num_cols].astype(float).round(4)
    order_signal = (
        slim.get("frequency_scaled_median", 0).fillna(0)
        + slim.get("monetary_scaled_median", 0).fillna(0)
        - slim.get("recency_days_scaled_median", 0).fillna(0)
    )
    return slim.assign(_order=order_signal).sort_values("_order", ascending=False).drop(columns="_order")

def name_clusters(profile_df: pd.DataFrame, cluster_col: str) -> dict:
    cols = {"freq":"frequency_scaled_median","mon":"monetary_scaled_median","rec":"recency_days_scaled_median"}
    med = profile_df.set_index(cluster_col)[[v for v in cols.values() if v in profile_df.columns]].fillna(0)
    hv = med.get(cols["freq"], 0) + med.get(cols["mon"], 0) - med.get(cols["rec"], 0)
    ordering = hv.sort_values(ascending=False).index.tolist()
    names = {}
    if ordering:
        names[ordering[0]] = "Active High-Value"
        for c in ordering[1:]:
            names[c] = "Mid/Low-Value"
    return names

def pca_scatter(X: pd.DataFrame, labels: np.ndarray, title: str):
    pca = PCAAdapter(n_components=2, random_state=RANDOM_SEED)
    comps = pca.fit_transform(X); evr = getattr(pca, "explained_variance_ratio_", None)
    comps = np.asarray(comps)
    plt.figure(figsize=(7,6))
    plt.scatter(comps[:,0], comps[:,1], c=labels, s=6, alpha=0.6)
    if evr is not None: plt.title(f"{title} — PCA2 (var: {evr[0]:.2f}, {evr[1]:.2f})")
    else:               plt.title(f"{title} — PCA2")
    plt.xlabel("PC1"); plt.ylabel("PC2"); plt.grid(True); plt.show()

def save_model_artifacts(label: str, km, feature_names: list[str], out_dir: Path):
    meta = {
        "label": label,
        "backend": BACKEND,
        "n_clusters": int(getattr(km, "n_clusters", getattr(km.model, "n_clusters", None) or 0)),
        "random_state": RANDOM_SEED,
        "inertia": float(getattr(km, "inertia_", np.nan)) if getattr(km, "inertia_", None) is not None else None,
        "feature_names": feature_names,
        "created_at": datetime.now().isoformat(timespec="seconds"),
    }
    try:
        centers = km.model.cluster_centers_ if hasattr(km, "model") else None
    except Exception:
        centers = None

    (out_dir / "models").mkdir(exist_ok=True, parents=True)
    with open(out_dir / "models" / f"{label}_meta.json", "w", encoding="utf-8") as f:
        json.dump(meta, f, indent=2)
    if centers is not None:
        np.savez_compressed(out_dir / "models" / f"{label}_centers.npz", centers=np.asarray(centers))
    if BACKEND == "cpu":
        try:
            import joblib
            joblib.dump(km.model, out_dir / "models" / f"{label}.joblib")
        except Exception as e:
            print(f"[WARN] Could not save sklearn joblib model ({label}): {e}")

# Train final models with confirmed ks
km_full, labels_full, sil_full, db_full, ch_full = fit_kmeans_and_report(X_full, final_k_full, "FULL", n_init=50, max_iter=300)
km_rfm3, labels_rfm3, sil_rfm3, db_rfm3, ch_rfm3 = fit_kmeans_and_report(X_rfm3, final_k_rfm3, "RFM3", n_init=50, max_iter=300)

# Attach labels and show cluster sizes
df_full_labels = attach_labels(cust_ids_full, labels_full, "cluster_full")
df_rfm3_labels = attach_labels(cust_ids_rfm3, labels_rfm3, "cluster_rfm3")
df_labels = df_full_labels.merge(df_rfm3_labels, on=id_col, how="inner")

print("\nCluster sizes (counts):")
print("FULL:", df_labels["cluster_full"].value_counts().sort_index().to_dict())
print("RFM3:", df_labels["cluster_rfm3"].value_counts().sort_index().to_dict())

# Load raw feature tables for profiling
feat_full_df = pd.read_parquet(DATA_DIR_IN / "model_features_customers_full.parquet")
feat_rfm3_df = pd.read_parquet(DATA_DIR_IN / "model_features_customers_rfm3.parquet")

feat_full_labeled = feat_full_df.merge(df_labels[[id_col, "cluster_full"]], on=id_col, how="left")
feat_rfm3_labeled = feat_rfm3_df.merge(df_labels[[id_col, "cluster_rfm3"]], on=id_col, how="left")

profile_full = cluster_profile(feat_full_labeled, "cluster_full")
profile_rfm3 = cluster_profile(feat_rfm3_labeled, "cluster_rfm3")

exec_full = executive_profile(profile_full, "cluster_full")
exec_rfm3 = executive_profile(profile_rfm3, "cluster_rfm3")

names_full = name_clusters(profile_full, "cluster_full")
names_rfm3 = name_clusters(profile_rfm3, "cluster_rfm3")

df_labels["cluster_full_name_model"] = df_labels["cluster_full"].map(names_full)  # keep model-based names
df_labels["cluster_rfm3_name"] = df_labels["cluster_rfm3"].map(names_rfm3)

print("\nSegment names (model-based):")
print("FULL:", names_full)
print("RFM3:", names_rfm3)

# ===== FIX (B): Detect HV index and create standardized binary label 0=HV, 1=ML
def _hv_signal(exec_df, cluster_col="cluster_full"):
    f = exec_df.set_index(cluster_col)
    return (
        f.get("frequency_scaled_median", 0).fillna(0)
        + f.get("monetary_scaled_median", 0).fillna(0)
        - f.get("recency_days_scaled_median", 0).fillna(0)
    )

hv_idx_full = int(_hv_signal(exec_full, "cluster_full").sort_values(ascending=False).index[0])
print(f"[naming] Detected High-Value cluster index (FULL): {hv_idx_full}")

df_labels["cluster_full_hv01"] = np.where(df_labels["cluster_full"] == hv_idx_full, 0, 1)
name_map_hv01 = {0: "High-Value", 1: "Mid/Low-Value"}
df_labels["cluster_full_name"] = df_labels["cluster_full_hv01"].map(name_map_hv01)  # standardized human name

# PCA plots
pca_scatter(X_rfm3, labels_rfm3, f"RFM3 clustering (k={final_k_rfm3})")
pca_scatter(X_full,  labels_full,  f"FULL clustering (k={final_k_full})")

# ===== FIX (C): Revenue by cluster (raw $), share, and lifts (vs global $/customer)
cust_base = pd.read_parquet(DATA_DIR_IN / "model_input_customers.parquet")[["CustomerID","monetary","frequency","recency_days"]]
df_rev = cust_base.merge(df_labels[[id_col, "cluster_full"]], on=id_col, how="inner")

rev_by_seg = (df_rev.groupby("cluster_full", as_index=False)
                     .agg(revenue_usd=("monetary","sum"),
                          n_customers=("CustomerID","nunique"),
                          freq_mean=("frequency","mean"),
                          recency_days_median=("recency_days","median")))

rev_total = float(rev_by_seg["revenue_usd"].sum())
rev_by_seg["revenue_share"] = rev_by_seg["revenue_usd"] / rev_total

global_mon_per_cust = df_rev["monetary"].mean()
rev_by_seg["lift_mon_per_cust"] = (rev_by_seg["revenue_usd"]/rev_by_seg["n_customers"] / global_mon_per_cust) - 1.0

REV_DIR = OUT_DIR / "revenue_profiles"; REV_DIR.mkdir(parents=True, exist_ok=True)
rev_by_seg.to_csv(REV_DIR / "revenue_by_cluster_full.csv", index=False)
print("[export] revenue_by_cluster_full.csv written")

# Save artifacts & exports (labels include cluster_full_hv01)
save_model_artifacts("kmeans_full", km_full,  [c for c in feat_full_df.columns if c.endswith("_scaled")], OUT_DIR)
save_model_artifacts("kmeans_rfm3", km_rfm3,  [c for c in feat_rfm3_df.columns if c.endswith("_scaled")], OUT_DIR)

out_labels_pq  = OUT_DIR / "customers_labeled_clusters.parquet"
out_labels_csv = OUT_DIR / "customers_labeled_clusters.csv"
df_labels.to_parquet(out_labels_pq, index=False)
df_labels.to_csv(out_labels_csv, index=False)

profile_full.to_csv(OUT_DIR / "cluster_profiles_full.csv", index=False)
profile_rfm3.to_csv(OUT_DIR / "cluster_profiles_rfm3.csv", index=False)
exec_full.to_csv(OUT_DIR / "cluster_profiles_full_executive.csv", index=False)
exec_rfm3.to_csv(OUT_DIR / "cluster_profiles_rfm3_executive.csv", index=False)

print("\n=== Final Model Quality ===")
print(f"FULL  (k={final_k_full}): silhouette={sil_full:.4f}, DB={db_full:.4f}, CH={ch_full:,.1f}")
print(f"RFM3  (k={final_k_rfm3}): silhouette={sil_rfm3:.4f}, DB={db_rfm3:.4f}, CH={ch_rfm3:,.1f}")

print("\n=== Outputs written ===")
print(f"- Labels (incl. hv01): {out_labels_pq}  |  {out_labels_csv}")
print(f"- Executive (FULL):    {OUT_DIR / 'cluster_profiles_full_executive.csv'}")
print(f"- Executive (RFM3):    {OUT_DIR / 'cluster_profiles_rfm3_executive.csv'}")
print(f"- Revenue profiles:    {REV_DIR / 'revenue_by_cluster_full.csv'}")
print(f"- Model artifacts dir: {OUT_DIR / 'models'}")

print("\n✅ Block 3 complete")
```

```python
# ============================================
# Block 5 — Bootstrapping Stability (self-contained, auto-k)
# Requires: X_full, X_rfm3, cust_ids_full, cust_ids_rfm3, and final_k_* from Blocks 1–3
# ============================================

import json, warnings, itertools
from pathlib import Path
from datetime import datetime
import numpy as np
import pandas as pd
warnings.filterwarnings("ignore")

# --- Fallbacks if adapters aren't already in memory
if 'silhouette_score_adapter' not in globals():
    from sklearn.metrics import silhouette_score as _sk_sil
    def silhouette_score_adapter(X, labels):
        return float(_sk_sil(np.asarray(X, np.float32), labels))

if 'KMeansAdapter' not in globals():
    from sklearn.cluster import KMeans as _skKMeans
    class KMeansAdapter:
        def __init__(self, n_clusters, n_init=30, max_iter=300, random_state=42):
            self.backend = "cpu"
            self.model = _skKMeans(n_clusters=n_clusters, n_init=n_init, max_iter=max_iter, random_state=random_state)
            self.inertia_ = None
        def fit_predict(self, X):
            Xnp = np.asarray(X, np.float32)
            labels = self.model.fit_predict(Xnp)
            self.inertia_ = float(self.model.inertia_)
            return labels

from sklearn.metrics import adjusted_rand_score, normalized_mutual_info_score
from sklearn.metrics.cluster import contingency_matrix

# --- Paths
DATA_DIR_IN = Path("clean/model_input")
OUT_DIR     = DATA_DIR_IN / "clusters"
VAL_DIR     = OUT_DIR / "validation"
VAL_DIR.mkdir(parents=True, exist_ok=True)

# --- Hungarian 
def try_hungarian_align(y_ref: np.ndarray, y_pred: np.ndarray, k: int) -> np.ndarray:
    try:
        from scipy.optimize import linear_sum_assignment
        C = contingency_matrix(y_ref, y_pred, sparse=False)
        cost = C.max() - C
        ri, ci = linear_sum_assignment(cost)
        mapping = {ci[j]: ri[j] for j in range(len(ri))}
        return np.vectorize(lambda t: mapping.get(t, t))(y_pred)
    except Exception:
        labels = list(range(k))
        best_map, best_agree = None, -1.0
        for perm in itertools.permutations(labels):
            m = {labels[j]: perm[j] for j in range(k)}
            aligned = np.vectorize(lambda t: m.get(t, t))(y_pred)
            agree = (aligned == y_ref).mean()
            if agree > best_agree: best_agree, best_map = agree, m
        return np.vectorize(lambda t: best_map.get(t, t))(y_pred)

# --- Bootstrap runner
def boot_kmeans_stability(X, ids_df, base_labels, dataset_name, k, n_boot=30, sample_frac=0.80, n_init=30, max_iter=300, seed=42):
    id_col = "CustomerID"
    X_id = X.join(ids_df).set_index(id_col)
    all_ids = X_id.index.to_numpy()

    rng = np.random.default_rng(seed)
    base_map = pd.Series(base_labels, index=ids_df[id_col].to_numpy(), name="base").to_dict()

    rows = []
    for b in range(n_boot):
        sub_seed = int(rng.integers(0, 1_000_000))
        rng_b = np.random.default_rng(sub_seed)
        n_sample = int(len(all_ids) * sample_frac)
        boot_ids = rng_b.choice(all_ids, size=n_sample, replace=False)
        Xb = X_id.loc[boot_ids]

        km = KMeansAdapter(n_clusters=k, n_init=n_init, max_iter=max_iter, random_state=sub_seed)
        y_pred = km.fit_predict(Xb)
        y_ref  = np.array([base_map[cid] for cid in boot_ids], dtype=int)
        y_algn = try_hungarian_align(y_ref, y_pred, k)

        agree = float((y_algn == y_ref).mean())
        ari   = float(adjusted_rand_score(y_ref, y_algn))
        nmi   = float(normalized_mutual_info_score(y_ref, y_algn))
        counts = pd.Series(y_algn).value_counts(normalize=True).sort_index()

        rows.append({
            "boot": b+1, "seed": sub_seed, "n_rows": int(n_sample),
            "agreement": agree, "ARI": ari, "NMI": nmi,
            **{f"share_cluster_{i}": float(counts.get(i, 0.0)) for i in range(k)}
        })

    df_boot = pd.DataFrame(rows)
    summary = {
        "dataset": dataset_name, "k": int(k), "n_boot": int(n_boot), "sample_frac": float(sample_frac),
        "agreement": {"mean": float(df_boot["agreement"].mean()), "median": float(df_boot["agreement"].median()),
                      "p10": float(df_boot["agreement"].quantile(0.10)), "p90": float(df_boot["agreement"].quantile(0.90))},
        "ARI": {"mean": float(df_boot["ARI"].mean()), "median": float(df_boot["ARI"].median()),
                "p10": float(df_boot["ARI"].quantile(0.10)), "p90": float(df_boot["ARI"].quantile(0.90))},
        "NMI": {"mean": float(df_boot["NMI"].mean()), "median": float(df_boot["NMI"].median()),
                "p10": float(df_boot["NMI"].quantile(0.10)), "p90": float(df_boot["NMI"].quantile(0.90))},
        "acceptance_rule": "Stable if agreement_median >= 0.90 or ARI_median >= 0.80",
    }

    csv_path  = VAL_DIR / f"bootstrap_{dataset_name.lower()}_k{k}.csv"
    json_path = VAL_DIR / f"bootstrap_{dataset_name.lower()}_k{k}_summary.json"
    df_boot.to_csv(csv_path, index=False)
    with open(json_path, "w", encoding="utf-8") as f: json.dump(summary, f, indent=2)

    print(f"[saved] {csv_path.name} & {json_path.name}")
    print(pd.DataFrame({
        "metric":["agreement","ARI","NMI"],
        "mean":[summary["agreement"]["mean"], summary["ARI"]["mean"], summary["NMI"]["mean"]],
        "median":[summary["agreement"]["median"], summary["ARI"]["median"], summary["NMI"]["median"]],
        "p10":[summary["agreement"]["p10"], summary["ARI"]["p10"], summary["NMI"]["p10"]],
        "p90":[summary["agreement"]["p90"], summary["ARI"]["p90"], summary["NMI"]["p90"]],
    }).round(4))
    print("Decision (rule):", "✅ STABLE" if (summary["agreement"]["median"] >= 0.90 or summary["ARI"]["median"] >= 0.80) else "⚠️ REVIEW")
    return df_boot

# --- Auto-k 
K_STABILITY_FULL  = int(globals().get("final_k_full", 2))
K_STABILITY_RFM3  = int(globals().get("final_k_rfm3", 2))
print(f"[bootstrap] FULL uses k={K_STABILITY_FULL}")
print(f"[bootstrap] RFM3 uses k={K_STABILITY_RFM3}")

# --- Base labels
if 'df_full_labels' in globals():
    base_full = df_full_labels['cluster_full'].to_numpy()
else:
    base_full = pd.read_csv(OUT_DIR/'customers_labeled_clusters.csv')['cluster_full'].to_numpy()

try:
    if 'df_rfm3_labels' in globals() and 'cluster_rfm3' in df_rfm3_labels.columns:
        base_rfm3 = df_rfm3_labels['cluster_rfm3'].to_numpy()
    else:
        base_rfm3 = pd.read_csv(OUT_DIR/'customers_labeled_clusters.csv')['cluster_rfm3'].to_numpy()
except Exception:
    base_rfm3 = None  

# --- Run
dfb_full = boot_kmeans_stability(X_full,  cust_ids_full,  base_full, "FULL",  k=K_STABILITY_FULL)
if base_rfm3 is not None:
    dfb_rfm3 = boot_kmeans_stability(X_rfm3,  cust_ids_rfm3,  base_rfm3, "RFM3",  k=K_STABILITY_RFM3)
else:
    print("[bootstrap] Skipped RFM3 (no labels)")
```

```python
# ============================================
# Block 6A — Crosstab & Consistency between FULL_k2 and RFM3_k2
# ============================================

import json, itertools, warnings
from pathlib import Path
import numpy as np
import pandas as pd
from sklearn.metrics import adjusted_rand_score, normalized_mutual_info_score
from sklearn.metrics.cluster import contingency_matrix

warnings.filterwarnings("ignore")

DATA_DIR_IN = Path("clean/model_input")
OUT_DIR     = DATA_DIR_IN / "clusters"
VAL_DIR     = OUT_DIR / "validation"
VAL_DIR.mkdir(parents=True, exist_ok=True)

id_col = "CustomerID"
if "df_labels" not in globals():
    df_full_labels = cust_ids_full.copy();  df_full_labels["cluster_full"] = labels_full
    df_rfm3_labels = cust_ids_rfm3.copy();  df_rfm3_labels["cluster_rfm3"] = labels_rfm3
    df_labels = df_full_labels.merge(df_rfm3_labels, on=id_col, how="inner")

ct_counts = pd.crosstab(df_labels["cluster_full"], df_labels["cluster_rfm3"])
ct_rowpct = pd.crosstab(df_labels["cluster_full"], df_labels["cluster_rfm3"], normalize="index").round(4)

print("\n=== Crosstab: FULL_k2 (rows) vs RFM3_k2 (cols) — counts ===")
display(ct_counts)
print("\n=== Crosstab: FULL_k2 vs RFM3_k2 — row % ===")
display(ct_rowpct)

def optimal_map_hungarian(y_ref, y_pred, k=2):
    try:
        from scipy.optimize import linear_sum_assignment
        C = contingency_matrix(y_ref, y_pred, sparse=False)
        cost = C.max() - C
        ri, ci = linear_sum_assignment(cost)
        return {ci[j]: ri[j] for j in range(len(ri))}
    except Exception:
        labels = list(range(k))
        best_map, best_agree = None, -1
        for perm in itertools.permutations(labels):
            m = {labels[j]: perm[j] for j in range(k)}
            aligned = np.vectorize(lambda t: m.get(t, t))(y_pred)
            agree = (aligned == y_ref).mean()
            if agree > best_agree: best_agree, best_map = agree, m
        return best_map

y_r = df_labels["cluster_rfm3"].to_numpy()
y_f = df_labels["cluster_full"].to_numpy()
mapping_full_to_rfm3 = optimal_map_hungarian(y_r, y_f, k=2)
y_f_aligned = np.vectorize(lambda t: mapping_full_to_rfm3.get(t, t))(y_f)

agreement = float((y_f_aligned == y_r).mean())
ari       = float(adjusted_rand_score(y_r, y_f_aligned))
nmi       = float(normalized_mutual_info_score(y_r, y_f_aligned))
print("\n=== Consistency metrics (FULL aligned → RFM3) ===")
print(f"Agreement: {agreement:.4f} | ARI: {ari:.4f} | NMI: {nmi:.4f}")
print(f"Optimal label mapping (FULL → RFM3): {mapping_full_to_rfm3}")

ct_counts.to_csv(VAL_DIR / "crosstab_fullk2_vs_rfm3k2_counts.csv")
ct_rowpct.to_csv(VAL_DIR / "crosstab_fullk2_vs_rfm3k2_rowpct.csv")
with open(VAL_DIR / "full_to_rfm3_labelmap.json", "w", encoding="utf-8") as f:
    json.dump({int(k): int(v) for k, v in mapping_full_to_rfm3.items()}, f, indent=2)

print("\nSaved: crosstab counts/rowpct + full_to_rfm3_labelmap.json")
```

```python
# ============================================
# Block 6B — Disagreement Drill-down (FULL_k2 vs RFM3_k2)
# ============================================

import math
from pathlib import Path
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

DATA_DIR_IN = Path("clean/model_input")
OUT_DIR     = DATA_DIR_IN / "clusters"
VAL_DIR     = OUT_DIR / "validation"
VAL_DIR.mkdir(parents=True, exist_ok=True)

id_col = "CustomerID"
assert "df_labels" in globals(), "Run Blocks 3/6A first."

feat_full_df = pd.read_parquet(DATA_DIR_IN / "model_features_customers_full.parquet")
feat_rfm3_df = pd.read_parquet(DATA_DIR_IN / "model_features_customers_rfm3.parquet")
df = feat_full_df.merge(df_labels[[id_col,"cluster_full","cluster_rfm3"]], on=id_col, how="inner")

def label_pair(row):
    f, r = int(row["cluster_full"]), int(row["cluster_rfm3"])
    if f==0 and r==0: return "match_00"
    if f==1 and r==1: return "match_11"
    if f==0 and r==1: return "flip_F0_to_R1"
    if f==1 and r==0: return "flip_F1_to_R0"
    return "other"
df["pair"] = df.apply(label_pair, axis=1)

sizes = df["pair"].value_counts().to_frame("count")
sizes["share"] = sizes["count"]/len(df)
print("=== Cohort sizes ==="); display(sizes)

rfm_feats = [c for c in ["recency_days_scaled","frequency_scaled","monetary_scaled"] if c in df.columns]
extra_feats = [c for c in ["avg_ticket_scaled","pct_discounted_scaled","n_categories_scaled"] if c in df.columns]
all_feats = rfm_feats + extra_feats
print("\nFeatures available for drill-down:")
print("RFM:", rfm_feats); print("EXTRA:", extra_feats)

def summarize_by_cohort(df_in, features, cohort_col="pair"):
    agg = df_in.groupby(cohort_col)[features].agg(["mean","median","std","min","max","count"])
    agg.columns = [f"{f}_{s}" for f,s in agg.columns.to_flat_index()]
    return agg.reset_index()

summary = summarize_by_cohort(df, all_feats, "pair")
print("\n=== Cohort summary (RFM + EXTRA) ==="); display(summary.round(4))

def cohens_d(x, y):
    x = np.asarray(x, float); y = np.asarray(y, float)
    n1, n2 = len(x), len(y)
    if n1<2 or n2<2: return np.nan
    s1, s2 = np.nanstd(x, ddof=1), np.nanstd(y, ddof=1)
    if (s1==0 and s2==0) or np.isnan(s1) or np.isnan(s2): return np.nan
    sp = math.sqrt(((n1-1)*s1**2 + (n2-1)*s2**2)/(n1+n2-2))
    if sp==0: return np.nan
    return (np.nanmean(x) - np.nanmean(y)) / sp

def compare_flip_vs_match(df_all, flip_name, match_name, features):
    a = df_all[df_all["pair"]==flip_name]; b = df_all[df_all["pair"]==match_name]
    rows=[]
    for f in features:
        rows.append({
            "feature": f, "flip": flip_name, "match": match_name,
            "mean_flip": a[f].mean(), "mean_match": b[f].mean(), "mean_diff": a[f].mean()-b[f].mean(),
            "median_flip": a[f].median(), "median_match": b[f].median(), "median_diff": a[f].median()-b[f].median(),
            "cohens_d": cohens_d(a[f].values, b[f].values), "n_flip": len(a), "n_match": len(b)
        })
    return pd.DataFrame(rows)

cmp_F0toR1 = compare_flip_vs_match(df, "flip_F0_to_R1", "match_00", all_feats)
cmp_F1toR0 = compare_flip_vs_match(df, "flip_F1_to_R0", "match_11", all_feats)

def show_effects(cmp_df, title):
    out = cmp_df.copy()
    out["abs_d"] = out["cohens_d"].abs()
    out = out.sort_values("abs_d", ascending=False)
    print(f"\n=== Effect sizes — {title} ===")
    display(out[["feature","mean_flip","mean_match","mean_diff","median_flip","median_match","median_diff","cohens_d","n_flip","n_match"]].round(4))
    return out

eff_F0toR1 = show_effects(cmp_F0toR1, "flip_F0_to_R1 vs match_00")
eff_F1toR0 = show_effects(cmp_F1toR0, "flip_F1_to_R0 vs match_11")

def quick_boxplots(df_all, eff_tbl, flip_name, match_name, top_n=2):
    feats = eff_tbl.sort_values("cohens_d", key=lambda s: s.abs(), ascending=False)["feature"].head(top_n).tolist()
    for f in feats:
        plt.figure(figsize=(6,4))
        data = [df_all.loc[df_all["pair"]==flip_name, f].values,
                df_all.loc[df_all["pair"]==match_name, f].values]
        plt.boxplot(data, labels=[flip_name, match_name], showfliers=False)
        plt.title(f"{f} — {flip_name} vs {match_name}")
        plt.ylabel(f); plt.grid(True, axis="y", alpha=0.3); plt.show()

quick_boxplots(df, eff_F0toR1, "flip_F0_to_R1", "match_00", top_n=2)
quick_boxplots(df, eff_F1toR0, "flip_F1_to_R0", "match_11", top_n=2)

sizes.to_csv(VAL_DIR / "cohort_sizes_full_vs_rfm3.csv")
summary.to_csv(VAL_DIR / "cohort_summary_rfm_extra.csv", index=False)
eff_F0toR1.to_csv(VAL_DIR / "effects_flip_F0_to_R1_vs_match_00.csv", index=False)
eff_F1toR0.to_csv(VAL_DIR / "effects_flip_F1_to_R0_vs_match_11.csv", index=False)
df[df["pair"]=="flip_F0_to_R1"][[id_col]].to_csv(VAL_DIR / "ids_flip_F0_to_R1.csv", index=False)
df[df["pair"]=="flip_F1_to_R0"][[id_col]].to_csv(VAL_DIR / "ids_flip_F1_to_R0.csv", index=False)

print("\nSaved cohort/effects/ids CSVs under:", VAL_DIR)
```

```python
# ============================================
# Block 6C — Harmonized label & Tableau exports (softer guardrails)
# ============================================

import json
import warnings
from pathlib import Path
import numpy as np
import pandas as pd

warnings.filterwarnings("ignore")

DATA_DIR_IN = Path("clean/model_input")
OUT_DIR     = DATA_DIR_IN / "clusters"
TAB_DIR     = OUT_DIR / "tableau"
TAB_DIR.mkdir(parents=True, exist_ok=True)

id_col = "CustomerID"

# ---------- Load features (FULL) ----------
feat_full_df = pd.read_parquet(DATA_DIR_IN / "model_features_customers_full.parquet")

# ---------- Ensure labels ----------
# Prefer labels already in memory from Block 3; otherwise load from saved CSV.
if "df_labels" in globals():
    pass
else:
    labels_path = OUT_DIR / "customers_labeled_clusters.csv"
    assert labels_path.exists(), (
        "labels dataframe not in memory and customers_labeled_clusters.csv not found. "
        "Run Block 3 first or provide the labels CSV."
    )
    tmp = pd.read_csv(labels_path)
    need = [id_col, "cluster_full", "cluster_rfm3", "cluster_full_hv01"]
    need = [c for c in need if c in tmp.columns]  # tolerate older CSVs
    df_labels = tmp[need].copy()

# ---------- Merge features + labels (bring hv01 if available) ----------
use_cols = [id_col, "cluster_full", "cluster_rfm3", "cluster_full_hv01"]
use_cols = [c for c in use_cols if c in df_labels.columns]
dfh = feat_full_df.merge(df_labels[use_cols], on=id_col, how="inner")

# If hv01 is missing (older CSV), fallback: assume 0=HV (less robust) — better re-run Block 3
if "cluster_full_hv01" not in dfh.columns:
    dfh["cluster_full_hv01"] = (dfh["cluster_full"] != 0).astype(int)

# ---------- Agreement / disagreement flag (uses raw numeric labels just for reference) ----------
def _pair(row):
    f, r = int(row["cluster_full"]), int(row["cluster_rfm3"])
    if f == 0 and r == 0: return "match_00"
    if f == 1 and r == 1: return "match_11"
    if f == 0 and r == 1: return "flip_F0_to_R1"
    if f == 1 and r == 0: return "flip_F1_to_R0"
    return "other"
dfh["pair"] = dfh.apply(_pair, axis=1)

# ---------- Softer guardrails (conservative) — on standardized hv01 labels ----------
# Downgrade only when BOTH freq and monetary are low.
THRESH_DOWN_FREQ = 0.40
THRESH_DOWN_MON  = 0.40

# Promote when freq and monetary are jointly strong.
THRESH_UP_FREQ_1 = 0.60
THRESH_UP_MON_1  = 0.45
THRESH_UP_FREQ_2 = 0.50
THRESH_UP_MON_2  = 0.55

# Available features
rfm_feats   = [c for c in ["recency_days_scaled", "frequency_scaled", "monetary_scaled"] if c in dfh.columns]
extra_feats = [c for c in ["avg_ticket_scaled", "pct_discounted_scaled", "n_categories_scaled"] if c in dfh.columns]
all_feats   = rfm_feats + extra_feats

# Harmonized segment starting from standardized FULL label (0=HV, 1=ML)
seg_h = dfh["cluster_full_hv01"].astype(int).copy()

mask_down = (
    dfh["cluster_full_hv01"].eq(0) &
    (dfh["frequency_scaled"] < THRESH_DOWN_FREQ) &
    (dfh["monetary_scaled"]  < THRESH_DOWN_MON)
)

mask_up = (
    dfh["cluster_full_hv01"].eq(1) & (
        ((dfh["frequency_scaled"] >= THRESH_UP_FREQ_1) & (dfh["monetary_scaled"] >= THRESH_UP_MON_1)) |
        ((dfh["frequency_scaled"] >= THRESH_UP_FREQ_2) & (dfh["monetary_scaled"] >= THRESH_UP_MON_2))
    )
)

seg_h[mask_down] = 1
seg_h[mask_up]   = 0

dfh["seg_harmonized"] = seg_h.astype(int)
name_map = {0: "High-Value", 1: "Mid/Low-Value"}
dfh["seg_harmonized_name"] = dfh["seg_harmonized"].map(name_map)

# Human-readable names (FULL: from hv01; RFM3: numeric assumption 0=HV,1=ML if you kept k=2 there)
dfh["cluster_full_name"]   = dfh["cluster_full_hv01"].map(name_map)
dfh["cluster_rfm3_name"]   = dfh["cluster_rfm3"].map(name_map) if "cluster_rfm3" in dfh.columns else np.nan

# ---------- Quick QA printouts ----------
counts = dfh["seg_harmonized"].value_counts().sort_index()
print("Harmonized counts:", counts.to_dict())
print("Harmonized share:", (counts / len(dfh)).round(4).to_dict())
print("Downgraded HV:", int(mask_down.sum()))
print("Promoted HV:",   int(mask_up.sum()))

def make_lifts(df: pd.DataFrame, seg_col: str, feats: list[str]) -> pd.DataFrame:
    """
    Compute per-segment lifts versus the global mean for the given features.
    Returns a tidy table: segment, feature, segment_mean, global_mean, lift, lift_pct, n_rows.
    lift = (segment_mean / global_mean) - 1
    """
    global_means = df[feats].mean().replace(0.0, np.nan)
    rows = []
    for seg_value, grp in df.groupby(seg_col, dropna=False):
        seg_means = grp[feats].mean()
        tmp = pd.DataFrame({
            "segment": seg_value,
            "feature": feats,
            "segment_mean": seg_means.values,
            "global_mean": global_means.loc[feats].values,
            "n_rows": len(grp)
        })
        tmp["lift"] = (tmp["segment_mean"] / tmp["global_mean"]) - 1.0
        tmp["lift_pct"] = tmp["lift"] * 100.0
        tmp.replace([np.inf, -np.inf], np.nan, inplace=True)
        tmp.fillna(0.0, inplace=True)
        rows.append(tmp)
    out = pd.concat(rows, ignore_index=True)
    out = out.sort_values(["segment", "lift"], ascending=[True, False]).reset_index(drop=True)
    return out

def make_profiles(df: pd.DataFrame, seg_col: str, feats: list[str]) -> pd.DataFrame:
    prof = df.groupby(seg_col, as_index=False)[feats].agg(
        ["mean", "median", "std", "min", "max", "count"]
    )
    new_cols = []
    for col in prof.columns.to_flat_index():
        if isinstance(col, tuple) and col[1] != "":
            base, stat = col
            new_cols.append(f"{base}_{stat}")
        else:
            new_cols.append(col if isinstance(col, str) else seg_col)
    prof.columns = new_cols
    sizes = (
        df[seg_col].value_counts().sort_index()
        .rename("count_rows").reset_index().rename(columns={"index": seg_col})
    )
    prof = prof.merge(sizes, on=seg_col, how="left")
    prof["share"] = prof["count_rows"] / len(df)
    return prof

profiles_h = make_profiles(dfh, "seg_harmonized", all_feats).sort_values("seg_harmonized").reset_index(drop=True)
lifts_h    = make_lifts(dfh, "seg_harmonized", all_feats)

# ---------- Tableau master dataset ----------
keep_cols = [
    id_col, "cluster_full", "cluster_full_hv01", "cluster_rfm3",
    "cluster_full_name", "cluster_rfm3_name",
    "pair", "seg_harmonized", "seg_harmonized_name", *all_feats
]
tableau_df = dfh[keep_cols].copy()

# ---------- Exports ----------
tableau_path  = TAB_DIR / "customers_segments_tableau.csv"
profiles_path = TAB_DIR / "segment_profiles_harmonized.csv"
lifts_path    = TAB_DIR / "segment_lifts_harmonized.csv"
dict_path     = TAB_DIR / "data_dictionary_segments.csv"

tableau_df.to_csv(tableau_path, index=False)
profiles_h.to_csv(profiles_path, index=False)
lifts_h.to_csv(lifts_path, index=False)

# Data dictionary (adds hv01)
data_dict = pd.DataFrame([
    (id_col, "ID", "Unique customer identifier", ""),
    ("cluster_full", "INT", "K-Means (FULL) label (raw index)", ""),
    ("cluster_full_hv01", "INT{0,1}", "Standardized FULL label", "0=High-Value, 1=Mid/Low-Value"),
    ("cluster_rfm3", "INT{0,1}", "K-Means (RFM3) label", "0=High-Value, 1=Mid/Low-Value"),
    ("cluster_full_name", "STRING", "Human-readable label (FULL, hv01)", ""),
    ("cluster_rfm3_name", "STRING", "Human-readable label (RFM3, numeric assumption)", ""),
    ("pair", "STRING", "FULL vs RFM3 agreement flag", "match_00, match_11, flip_F0_to_R1, flip_F1_to_R0"),
    ("seg_harmonized", "INT{0,1}", "Harmonized label (guardrails applied)", "0=High-Value, 1=Mid/Low-Value"),
    ("seg_harmonized_name", "STRING", "Human-readable harmonized label", ""),
] + [(f, "FLOAT[0,1]", "Scaled feature", "min-max scaled to [0,1]") for f in all_feats],
    columns=["field", "type", "description", "notes"]
)
data_dict.to_csv(dict_path, index=False)

print("\n=== Exports written ===")
print(f"- Tableau master: {tableau_path}")
print(f"- Profiles:       {profiles_path}")
print(f"- Lifts:          {lifts_path}")
print("- Data dictionary:", dict_path)
```

```python
# ============================================
# Block 7 — Alternative Models Benchmark
# ============================================

import os, json, warnings, math
from pathlib import Path
from datetime import datetime
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

warnings.filterwarnings("ignore")

DATA_DIR_IN = Path("clean/model_input")
OUT_DIR     = DATA_DIR_IN / "clusters"
VAL_DIR     = OUT_DIR / "validation"
ALT_DIR     = OUT_DIR / "alt_models"
for d in [VAL_DIR, ALT_DIR]: d.mkdir(parents=True, exist_ok=True)

assert "X_full" in globals() and "X_rfm3" in globals(), "Run Blocks 1–3 first."
assert "cust_ids_full" in globals() and "cust_ids_rfm3" in globals(), "Run Block 1 first."

try:
    import cuml; BACKEND = "gpu"
except Exception:
    BACKEND = "cpu"

try:
    silhouette_score_adapter
except NameError:
    from sklearn.metrics import silhouette_score as sk_silhouette
    def silhouette_score_adapter(X, labels):
        return float(sk_silhouette(np.asarray(X, dtype=np.float32), labels))

def cluster_profile(df: pd.DataFrame, cluster_col: str) -> pd.DataFrame:
    feats = [c for c in df.columns if c.endswith("_scaled")]
    agg = df.groupby(cluster_col, as_index=False)[feats].agg(["mean","median","std","min","max","count"]).reset_index()
    agg.columns = [f"{b}_{s}" if s else b for b,s in agg.columns.to_flat_index()]
    sizes = df[cluster_col].value_counts().sort_index().rename("count")
    agg = agg.merge(sizes, left_on=cluster_col, right_index=True)
    agg["share"] = agg["count"]/len(df); return agg

def executive_profile(profile_df: pd.DataFrame, cluster_col: str) -> pd.DataFrame:
    keep = [c for c in [
        "recency_days_scaled_median","frequency_scaled_median","monetary_scaled_median",
        "avg_ticket_scaled_median","pct_discounted_scaled_median","n_categories_scaled_median"
    ] if c in profile_df.columns]
    slim = profile_df[[cluster_col,*keep,"share","count"]].copy()
    num = [c for c in slim.columns if c != cluster_col]
    slim[num] = slim[num].astype(float).round(4)
    signal = (slim.get("frequency_scaled_median",0)+slim.get("monetary_scaled_median",0)-slim.get("recency_days_scaled_median",0))
    return slim.assign(_order=signal).sort_values("_order", ascending=False).drop(columns="_order")

# GMM
try:
    from cuml.mixture import GaussianMixture as cuGMM
except Exception:
    cuGMM = None
from sklearn.mixture import GaussianMixture as skGMM

def gmm_fit_predict(X, k, random_state=42, covariance_type="full", n_init=10):
    X_np = np.asarray(X, dtype=np.float32)
    if cuGMM is not None:
        try:
            model = cuGMM(n_components=k, covariance_type=covariance_type, n_init=n_init, random_state=random_state)
            model.fit(X_np); labels = model.predict(X_np); backend = "gpu"
            try: ll = float(model.score_samples(X_np).sum())
            except Exception: ll = float(model.score(X_np)*len(X_np))
            return model, labels, ll, backend
        except Exception:
            pass
    model = skGMM(n_components=k, covariance_type=covariance_type, n_init=n_init, random_state=random_state)
    labels = model.fit_predict(X_np); backend = "cpu"
    try: ll = float(model.score_samples(X_np).sum())
    except Exception: ll = float(model.score(X_np)*len(X_np))
    return model, labels, ll, backend

def gmm_params_count(k, d, covar="full"):
    if covar == "full": cov_params = k * d * (d + 1) / 2
    elif covar == "diag": cov_params = k * d
    else: cov_params = k * d * (d + 1) / 2
    return int((k - 1) + k*d + cov_params)

def aic_bic(loglik, n_params, n_obs):
    if np.isnan(loglik): return np.nan, np.nan
    aic = 2*n_params - 2*loglik
    bic = n_params*np.log(n_obs) - 2*loglik
    return float(aic), float(bic)

# Agglomerative (Ward) sampled + projection
from sklearn.cluster import AgglomerativeClustering as skAgglo
from sklearn.metrics import davies_bouldin_score, calinski_harabasz_score

def agglomerative_ward_sampled(X, ids_df, dataset_name, k_values=(2,3), max_sample=8000, min_sample=3000, random_state=42):
    rng = np.random.default_rng(random_state)
    n_total = len(X); X_np_full = np.asarray(X, dtype=np.float32)
    results = []

    n_sample = min(max_sample, n_total)
    sample_idx = rng.choice(n_total, size=n_sample, replace=False)
    Xs = X.iloc[sample_idx].reset_index(drop=True).astype(np.float32)
    ids_s = ids_df.iloc[sample_idx].reset_index(drop=True)

    for k in k_values:
        cur_n = n_sample
        while True:
            try:
                agg = skAgglo(n_clusters=k, linkage="ward")
                labels_s = agg.fit_predict(np.asarray(Xs, dtype=np.float32))
                sil = float(silhouette_score_adapter(np.asarray(Xs, np.float32), labels_s))
                db  = float(davies_bouldin_score(np.asarray(Xs, np.float32), labels_s))
                ch  = float(calinski_harabasz_score(np.asarray(Xs, np.float32), labels_s))

                centroids = [np.asarray(Xs[labels_s==c].mean(axis=0), dtype=np.float32) for c in sorted(np.unique(labels_s))]
                C = np.vstack(centroids).astype(np.float32)
                x2 = (X_np_full**2).sum(axis=1, keepdims=True)
                c2 = (C**2).sum(axis=1, keepdims=True).T
                d2 = x2 - 2*(X_np_full@C.T) + c2
                labels_full = np.argmin(d2, axis=1).astype(int)

                lab_col_s = f"cluster_agg_ward_k{k}_sample"
                lab_col_f = f"cluster_agg_ward_k{k}"
                ids_s.assign(**{lab_col_s: labels_s}).to_csv(ALT_DIR / f"{dataset_name.lower()}_{lab_col_s}.csv", index=False)
                ids_df.assign(**{lab_col_f: labels_full}).to_csv(ALT_DIR / f"{dataset_name.lower()}_{lab_col_f}.csv", index=False)

                feat_path = DATA_DIR_IN / ("model_features_customers_rfm3.parquet" if dataset_name=="RFM3" else "model_features_customers_full.parquet")
                feat_df = pd.read_parquet(feat_path).merge(ids_df.assign(**{lab_col_f: labels_full}), on="CustomerID", how="left")
                prof = cluster_profile(feat_df, lab_col_f); exec_tbl = executive_profile(prof, lab_col_f)
                prof.to_csv(ALT_DIR / f"{dataset_name.lower()}_{lab_col_f}_profiles.csv", index=False)
                exec_tbl.to_csv(ALT_DIR / f"{dataset_name.lower()}_{lab_col_f}_executive.csv", index=False)

                results.append({"dataset": dataset_name, "model":"Agglomerative_Ward_proj","backend":"cpu",
                                "k":k,"silhouette":sil,"davies_bouldin":db,"calinski_harabasz":ch,
                                "loglik":np.nan,"AIC":np.nan,"BIC":np.nan,
                                "notes":f"ward on sample (n={cur_n}) + nearest-centroid projection"})
                break
            except MemoryError:
                cur_n = max(int(cur_n*0.6), min_sample)
                if cur_n < min_sample:
                    results.append({"dataset":dataset_name,"model":"Agglomerative_Ward_proj","backend":"cpu",
                                    "k":k,"silhouette":np.nan,"davies_bouldin":np.nan,"calinski_harabasz":np.nan,
                                    "loglik":np.nan,"AIC":np.nan,"BIC":np.nan,"notes":"ward OOM even after downsizing"})
                    break
                sample_idx = rng.choice(n_total, size=cur_n, replace=False)
                Xs = X.iloc[sample_idx].reset_index(drop=True).astype(np.float32)
                ids_s = ids_df.iloc[sample_idx].reset_index(drop=True)
    return results

# DBSCAN grid (GPU if available)
try:
    from cuml.cluster import DBSCAN as cuDBSCAN
except Exception:
    cuDBSCAN = None
from sklearn.cluster import DBSCAN as skDBSCAN

def dbscan_grid(X, ids_df, dataset_name, eps_grid=None, min_samples=50):
    if eps_grid is None: eps_grid = [0.05,0.08,0.10,0.12,0.15]
    best_row = None
    for eps in eps_grid:
        try:
            if cuDBSCAN is not None:
                dbs = cuDBSCAN(eps=eps, min_samples=min_samples); labels = dbs.fit_predict(np.asarray(X, np.float32)); backend="gpu"
            else:
                dbs = skDBSCAN(eps=eps, min_samples=min_samples); labels = dbs.fit_predict(np.asarray(X, np.float32)); backend="cpu"
        except Exception:
            continue
        unique = np.unique(labels)
        if (unique>=0).sum() >= 2:
            Xnp = np.asarray(X, np.float32)
            sil = float(silhouette_score_adapter(Xnp, labels))
            db  = float(davies_bouldin_score(Xnp, labels))
            ch  = float(calinski_harabasz_score(Xnp, labels))
        else:
            sil=db=ch=np.nan
        row = {"dataset":dataset_name,"model":"DBSCAN","backend":backend,"k":int((labels>=0).sum()),
               "silhouette":sil,"davies_bouldin":db,"calinski_harabasz":ch,"loglik":np.nan,"AIC":np.nan,"BIC":np.nan,
               "eps":eps,"min_samples":min_samples}
        if (best_row is None) or (np.nan_to_num(sil, nan=-1) > np.nan_to_num(best_row["silhouette"], nan=-1)):
            best_row = row
        ids_df.assign(**{f"cluster_dbscan_eps{str(eps).replace('.','')}": labels}).to_csv(ALT_DIR / f"{dataset_name.lower()}_cluster_dbscan_eps{str(eps).replace('.','')}.csv", index=False)
    return [best_row] if best_row is not None else []

def run_alt_models_on(X, ids_df, dataset_name):
    print(f"\n=== Running alt models on {dataset_name} (n={len(X):,}, d={X.shape[1]}) ===")
    results=[]; d, n = X.shape[1], len(X)
    for k in [2,3]:
        model, labels, ll, backend = gmm_fit_predict(X, k, covariance_type="full", n_init=10)
        Xnp = np.asarray(X, np.float32)
        sil = float(silhouette_score_adapter(Xnp, labels))
        db  = float(davies_bouldin_score(Xnp, labels))
        ch  = float(calinski_harabasz_score(Xnp, labels))
        p   = gmm_params_count(k, d, covar="full")
        aic, bic = aic_bic(ll, p, n)
        results.append({"dataset":dataset_name,"model":f"GMM_full_k{k}","backend":backend,"k":k,
                        "silhouette":sil,"davies_bouldin":db,"calinski_harabasz":ch,
                        "loglik":ll,"AIC":aic,"BIC":bic,"notes":""})
        lab_col=f"cluster_gmm_k{k}"
        ids_df.assign(**{lab_col:labels}).to_csv(ALT_DIR/f"{dataset_name.lower()}_{lab_col}.csv", index=False)
        feat_path = DATA_DIR_IN / ("model_features_customers_rfm3.parquet" if dataset_name=="RFM3" else "model_features_customers_full.parquet")
        feat_df = pd.read_parquet(feat_path).merge(ids_df.assign(**{lab_col:labels}), on="CustomerID", how="left")
        prof = cluster_profile(feat_df, lab_col); exec_tbl = executive_profile(prof, lab_col)
        prof.to_csv(ALT_DIR/f"{dataset_name.lower()}_{lab_col}_profiles.csv", index=False)
        exec_tbl.to_csv(ALT_DIR/f"{dataset_name.lower()}_{lab_col}_executive.csv", index=False)

    results.extend(agglomerative_ward_sampled(X, ids_df, dataset_name, k_values=(2,3), max_sample=8000, min_sample=3000))
    results.extend(dbscan_grid(X, ids_df, dataset_name, eps_grid=[0.05,0.08,0.10,0.12,0.15], min_samples=50))
    return results

all_results=[]
for name, X, ids in [("FULL", X_full, cust_ids_full), ("RFM3", X_rfm3, cust_ids_rfm3)]:
    all_results.extend(run_alt_models_on(X, ids, name))

cmp_alt = pd.DataFrame(all_results).sort_values(["dataset","model","k"]).reset_index(drop=True)
cmp_path = ALT_DIR / "alt_models_comparison.csv"
cmp_alt.to_csv(cmp_path, index=False)
print("\n=== Alternative Models — Comparison Table (head) ===")
display(cmp_alt.head(30))
print("Saved:", cmp_path)
```
